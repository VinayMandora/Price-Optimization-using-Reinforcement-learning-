## To create the data files according to the product catergory

```python
import pandas as pd
import numpy as np

# Assuming your original dataset is loaded into a DataFrame called 'original_data'
# Make sure to replace 'original_data.csv' with the actual file path if needed.
data =pd.read_csv('C:\\Users\\ishwa\\Downloads\\krr dataset\\final\\retail_data.csv')
```

```python
# Convert 'PurchaseDate' to datetime format
data['PurchaseDate'] = pd.to_datetime(data['PurchaseDate'])

# Extract temporal features

data['PurchaseMonth'] = data['PurchaseDate'].dt.month
data['PurchaseQuarter'] = data['PurchaseDate'].dt.quarter
data['DayOfWeek'] = data['PurchaseDate'].dt.dayofweek
```

```python
dele=['CustomerID', 'MarketingExpenditure','AnnualIncome','SpendingScore','PaymentMethod','StoreID', 'PurchaseDate','DiscountPercent' ]
data=data.drop(columns=dele, axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'ProductCategory' is a categorical column in your dataset
le = LabelEncoder()
# Fit and transform the 'ProductCategory' column
data['ProductType'] = le.fit_transform(data['StoreLocation'])
```

```python
data.head()
```

```python
data=data.drop(columns='StoreLocation', axis=1)
```

```python
data=data.drop(columns='DiscountApplied', axis=1)
```

```python
import pandas as pd

# Assuming you have a DataFrame named 'df' with a column 'category' representing product categories
# and other columns for data.

# Example DataFrame

#df = pd.DataFrame(data)

# Get unique categories
categories = data['ProductCategory'].unique()

print(categories)
```

```python
df=data.sample(frac=0.001)
```

```python
# Create separate datasets for each category
category_datasets = {}
for category in categories:
    category_datasets[category] = df[df['ProductCategory'] == category].copy()

```

```python
# Save each dataset to a separate file
for category, dataset in category_datasets.items():
    file_name = f"{category}_dataset.csv"
    dataset.to_csv(file_name, index=False)
    print(f"Dataset for {category} saved to {file_name}")
```
